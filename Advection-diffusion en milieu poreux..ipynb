Nous allons construire l'opérateur H et puis calculer ses valeur propre grace à la fonction numpy : numpy.linalg.eig .

En suivant le même raisonnement qu'à la question 2, nous pouvons dire que :
$$ H =\frac{1}{4\Delta \xi^2}A + \frac{\xi}{4\Delta \xi} B  $$
De plus, $\xi = \Delta \xi * m $ où m est la ligne de la matrice (chez nous , m ira de 1 à 99 car nous avons supprimer les ligne 0 et 100).donc :
$$ H =\frac{1}{4\Delta \xi^2}A + \frac{m}{4} B $$

```python
import numpy as np
```

```python
def matriceH(de,N): # Nous construisons la matrice associé à l'opérateur H
    #elle depend de 'de' qui est le pas et de N la taille de la matrice
    #Nous construisons le premier terme de H, ce terme vaudra D1 + U1 + L1
    D1 = np.diag(( (-1/(2*(de**2))) ) * np.ones(N)) 
    U1 = np.diag( (1/(4*(de**2))) * np.ones(N-1), k = 1)
    L1 = np.diag( (1/(4*(de**2))) * np.ones(N-1), k = -1)
    #Nous construisons le 2e terme de H, il vaut U2 + L2
    U2 = np.diag( np.ones(N-1), k = 1)
    L2 = np.diag( -1 * np.ones(N-1), k = -1)
    H2 = U2 + L2
    for i in range (N-1): #chaque ligne de H2 doit être multiplié par un m différent
        H2[i] = H2[i] * ( (i+1)/4 )
    H = D1 + U1 + L1 + H2
    return H
```

```python
de = 0.05
H = matriceH(de,99)
H
```

```python
value, vector = np.linalg.eig(H) # np.linalg.eig donne la matrice de vecteur propre et la liste des valeur propre
#Nous ne prenons que la liste des valeur propre
```

```python
highest = [] # Nous allons prendre les 5 plus grandes valeurs de la liste des valeur propre de H
value1 = value.tolist() 
for j in range(5):
    control = value1[0]
    for i in range(len(value1)): 
        if value1[i] > control : 
            control = value1[i]
    highest.append(control)
    value1.remove(control)
highest
```

Nous obtenons bien les même valeur. Les nombres décimaux sont là car l'opérateur H est calculer grâce à une approximation.

Nous avons les équations : 
$$\left(\frac{Ra}{4t}\frac{\partial^2}{\partial \xi^2}-k^2\right)w = -k^2 c\\
\frac{\partial c}{\partial t}-\frac{1}{t}\left(\frac{1}{4}\frac{\partial^2}{\partial \xi^2}+\frac{\xi}{2}\frac{\partial}{\partial\xi}-\frac{k^2 t}{Ra}\right)c=\sqrt{\frac{Ra}{\pi t}}e^{-\xi^2}w
$$

En utilisant les opérateurs de la question 2, cela donne :
$$ w = -L^{-1}k^2 c\\
\frac{\partial c}{\partial t} =  \sqrt{\frac{Ra}{\pi t}}e^{-\xi^2}w + Mc
$$
ce qui nous donne : 
$$ \frac{\partial c}{\partial t} =  (-\sqrt{\frac{Ra}{\pi t}}e^{-\xi^2}L^{-1}k^2 + M)c $$
Nous pouvons maintenant appliquer rk4 sur c.

Dans le code qui suit, nous rennomerons $\xi$  en e et $\Delta \xi$ en de

```python
def operator_L(de,r,t,k , N): # Nous construisons l'opérateur L
    #il dépend de 'de' le pas, Ra que nous avons rennomé en r , du temps t , de la constante k et de la taille N 
    D = np.diag(( (-1*r/(2*t*(de**2))) - k**2 ) * np.ones(N))
    U = np.diag( (r/(4*t*(de**2))) * np.ones(N-1), k = 1)
    L = np.diag( (r/(4*t*(de**2))) * np.ones(N-1), k = -1)
    M = D + U + L
    return M
```

```python
def operator_M(de,r,t,k, N): # Nous construisons l'opérateur M .Nous ne l'avons pas multiplier par 1/t ici mais nous le ferons dans la fct f
    #il dépend de 'de' le pas, Ra que nous avons rennomé en r , du temps t , de la constante k et de la taille N 
    D1 = np.diag(( (-1/(2*(de**2))) - ( (k**2)*t/r ) ) * np.ones(N))
    U1 = np.diag( (1/(4*(de**2))) * np.ones(N-1), k = 1)
    L1 = np.diag( (1/(4*(de**2))) * np.ones(N-1), k = -1)
    U2 = np.diag( np.ones(N-1), k = 1)
    L2 = np.diag( -1 * np.ones(N-1), k = -1)
    H2 = U2 + L2
    for i in range (N-1):
        H2[i] = H2[i] * ( (i+1)/4 )
    M = D1 + U1 + L1 + H2
    return M
```

```python
def f(e,t,c,L,M,j): # c' = f(c(t),t)
    # f dépend de e,le temps t, le vecteur c, les opérateur L et M au temps correspondant et j qui est une variable d'itteration indiquer par quelle ligne de L et M on multiplie c
    x = np.dot((-np.sqrt(r/(np.pi*t)) * np.exp(-e**2) * (k**2) * L[j] +(1/t) * M[j]  ), c)
    return x
```

```python
def first_rk4(de,r , k , c0 , t0, stop, h): 
    # résout le systeme c' = f(c(t),t)
    #de est le pas de e, r est la constante Ra, k la constante k, c0 le vecteur au temps t0, t0 le temps initial, stop est le temps auquelle on s'arrête et est le pas du temps
    c = np.copy(c0) # Nous faisons une copie pour ne pas modifier c0
    t = t0
    # on calcule les coefficient en même temps selon la méthode de Runge-Kutta
    for i in range(int((stop - t0)/h)):
        L = np.linalg.inv(operator_L(de , r , t , k , 99)) # on calcule l'inverse de l'opérateur L au temps t grace à la fonction numpy
        M = operator_M(de, r , t , k , 99) # on calcule l'opérateur M au temps t
        k1_value = np.zeros(99) # Nous travaillons avec des vecteur, nous construisons alors des array pour sauvegarder nos valeur
        k2_value = np.zeros(99)
        k3_value = np.zeros(99)
        k4_value = np.zeros(99)
        for j in range(len(c0)): # nous calculons k1 pour chaque composante du vecteur c en prenant le bon e pour la ligne avec laquelle nous travaillons
            e = de * (j+1)
            k1 = h*f(e,t,c,L,M,j)
            k1_value[j] += k1
        for k in range(len(c0)):
            e = de * (k+1)
            k2 = h*f(e,t + h/2,c + k1_value/2,L,M,k)
            k2_value[k] += k2
        for l in range(len(c0)):
            e = de * (l+1)
            k3 = h*f(e,t + h/2,c + k2_value/2,L,M,l)
            k3_value[l] += k3
        for n in range (len(c0)):
            e = de * (n+1)
            k4 = h*f(e,t + h,c + k3_value,L,M,n)
            k4_value[n] += k4
        c += 1/6 * (k1_value + 2*k2_value + 2*k3_value + k4_value)
        t += h
    return c
```

```python
#Nous entrons nos conditions initials
de = 0.05
r = 500
k = 0.1
t0 = 10**(-5)
stop = 10**(-4)
h = 10**(-8)
c0 = np.load("init_cond_c.npy")
value_5 = first_rk4(de,r , k , c0 , t0, stop, h) # nous sauvegardons la solution final
```

```python
maximum = [] #Nous allons prendre la plus grande valeur de notre solution
control = value_5[0]
for i in range (len(value_5)):
    if value_5[i] > control : 
        control = value_5[i]
maximum.append(control)
maximum
    
```
